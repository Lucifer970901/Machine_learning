**Matplotlib** 

a Python library used for creating static, animated, and interactive visualizations. It is a versatile tool that allows users to generate a wide variety of plots, including line plots, scatter plots, bar charts, and histograms. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([0,30])
ypoints = np.array([0,240])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,'o') # return only the points on the graph
plt.show()
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
ypoints = np.array([3,5,7,11,12,5,7]) # if only y points are given it takes  default x points
plt.plot(ypoints)
plt.show()
```

```python
ypoints = np.array([1,3,5,10,5,7,8])
plt.plot(ypoints,marker = '*') #it shows the points as stars as mentioned in the marker
plt.show()
```

```python
plt.plot(ypoints,marker = 'o',markersize = '10',mec = 'r') 
plt.show()
```

markersize is used to change the size of marker points on the plot, 'mec' stands for marker edge color, here 'r' indicates red.

```python
plt.plot(ypoints,marker = '*',markersize = 20,mec ='b',mfc = 'r' ) #mfc  - marker face color - here its red, and edge color is blue
plt.show()
```

```python
plt.plot(ypoints,ls = 'dashed') #ls stands for - linestyle
plt.show()
```

```python
plt.plot(ypoints,color = 'b',ls ='dashed',marker = 'o',mfc = 'c',mec = 'r',markersize = 20 ) # c for ceyon, color - for the color of the line here
plt.show()
```

```python
plt.plot(ypoints,ls = 'dashed',linewidth = 10)
plt.show()
```

```python
x1 = np.array([0,1,2,3])
x2 = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title("sports watch data")
plt.xlabel('average pulse')
plt.ylabel('calorie burnage')
plt.grid(ls = '--',linewidth = 1,color = 'green')
plt.show()
```

grid function returns grids in the plot, if you want specific axis gird, provide axis = x - returns grid x or vertical lines

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.scatter(x,y,color = "red") # scatters the points on the plot
x = np.array([80,85,90,95,100,105,110,117,120,125])
y = np.array([77,250,260,270,260,290,300,310,320,330])
plt.scatter(x,y,color = "green")
plt.show()
```

viridis is the colormap library in python

```python
import random
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)
plt.scatter(x,y,c = z,cmap='viridis')
```

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

creates a 1D array named x with 250 elements, each drawn from a normal distribution. The normal distribution has a mean (location) of 170 and a standard deviation (scale) of 10.

```python
y = np.array([20,30,15,25])
mylabels = ["Apples","Bananas","Cherries","Dates"]
plt.pie(y)
plt.legend(mylabels)
```

```python
y = np.array([20,30,15,25])
myexplode = [0,0,0.1,0]
mylabels = ["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels = mylabels,explode = myexplode)
plt.show()
```

```python
plt.pie(y,labels = mylabels,explode = myexplode,shadow = True)
plt.show()
```

```python
y = np.array([25,35,35,15])
mycolors = ['hotpink','c','r','purple']
mylabels = ['apple','banana','strawberry','blackcurrent']
myexplode = [0.2,0,0,0]
plt.pie(y,labels = mylabels,shadow = True, explode = myexplode,colors = mycolors)
plt.show()
```

```python
mylabels = ["Apples","Bananas","Cherries","Dates"]
myexplode = [0,0,0.1,0]
plt.pie(y,labels = mylabels,shadow = True,explode = myexplode)
plt.legend(title = "Four fruits")
plt.show()
```
