**Seaborn** 

Built on top of matplotlib provides higher level of specialized interface creating statistical graphics

we will use certain default dataset, that seaborn offers for  data manipulation and visualization. including, tips, titanic, iris and fmri etc

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip',hue = 'sex',data = tips)
plt.xlabel('Total bill')
plt.ylabel('Tips')
plt.title('Relationship between total bill vs tips')
```

**hue** - adds 3rd dimension to statistical plots by mapping a categorical variable with different colors.

use **scatter()** function from matplotlib to see the scatter of the points across variables.

```python
plt.scatter(tips['total_bill'],tips['tip'])
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint',y = 'signal', hue = 'event',style = 'region',data = fmri)
```

**lineplot** - draw seperate lines for each unique value or category in the event column and color them differently.

**style**  - adds another layer of differention for each categorical variables which uses different lines styles or marker styles. (like circles, square, triangle etc.)

**dashed = False**  - draws solid line for all categories.

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x = 'class', y = 'fare',hue = 'sex', data = titanic)
```

**Barplot** - shows the relationship between categorical variables and quntitative variables. which means quiantitiatve variables for each category of categorical variable.

**ci = None** - there will be no lines on the bars, if you set confidence interval to None.

```python
sns.barplot(x = 'class',y = 'fare',hue = 'sex',ci = None,data = titanic)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(x= 'petal_length',bins = 20,data = iris)
```

**histplot** - shows the distribution of numerical dataset, groups values into 'bins' then counts how many observations fall into bin. counts are represented with bars.


```python
sns.histplot(x = 'petal_length',kde = True,bins = 20,data = iris)
```

**kde = True** - Kernel density estimation, draws a curve on top of the bars.

**bins = 20** - here petal length values are divided into 20 equally sized intervals(bins).

```python
sns.kdeplot(x ='total_bill',data  = tips)
```

**kdeplot** - draws smooth curve based on the density estimate for the varible provided

```python
sns.kdeplot(hue = 'time',x = 'total_bill',fill = True,alpha = 0.6,linewidth = 1.5,data =tips)
```

**hue = 'time'** - draws 2 curves for variable based on the category of values it holds.

**fill = True** - fills the underneath the kde curve with a solid color.

**alpha = 0.6** - controls the transperency and opacity of the filled area, 0 means completely transperent and 1 means completely opaque.

**linewidth = 1.5** - sets the thickness of the kde curve line itself to 1.5 points

```python
sns.boxplot(x= 'day',y = 'total_bill',hue = 'time',palette = 'Set3',data = tips)
```

**boxplot** -  shows the distribution of the quantitative variable across different categories shows median, inter quartile ranges (IQR) whiskers - from the box edges show the range of most of the data. outside the whiskers the dots indicate outliers.

**palette = 'Set3'** - this parameter tells hue which color palette to use for hue variable.

```python
sns.violinplot(x = 'species',y = 'petal_length',data = iris)
```

**Voilinplot** - hybrid of box plot and kde plot, shows distribution of quantitative variable across different categories

if the plot os having wider body - indicates higher density of observations.

inside the violin, there are miniature box plot, the shape, multiple modes(peaks) and spread - shows distribution which box plot cannot show.

```python
sns.pairplot(data=iris,hue='species',diag_kind='hist')
plt.title('Iris Dataset Pairplot')
```

**pairplot** - generates grid plots that visualizes the distribution of each individual variable. diagonal shows the plot of variable against itself. off the diagonals shows scatter plots of each pair of variable allowing to see relationships.

```python
sns.jointplot(x='sepal_length', y='sepal_width',data=iris)
```

**Jointplot** - designed to display the relationship between 2 variables(bivariate analysis) simultaneously with their individual distributions(univariate analysis) on the same plot. 

this is used when you want to visualize the relationship between exactly 2 numerical variables. or to quickly assess the correlation, trends, and potential outliers in bivariate data.

```python
g = sns.FacetGrid(tips,col='day')

g.map(sns.histplot, 'total_bill')
```

**facet grid** - produces multiple plots in organized grid where each plot or facet represents subset of your data based on the levels of one or more categorical variables. it allows to visualize how relationship distribution changes across different categories.
