**Data pre processing with health dataset.**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../sample_data/health_data.csv.txt')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.to_string())
```

```python
x = df['Calories'].mean()
df.fillna({'Calories':x},inplace = True)
print(df.to_string())
```

```python
x = df['Calories'].mode()[0]
df.fillna({'Calories':x},inplace = True)
print(df.to_string())
```

```python
df['Date'] = pd.to_datetime(df['Date'],format = 'mixed')
print(df.to_string())
```

```python
df.loc[7,'Duration'] = 45
print(df.to_string())
```

```python
df.dropna(inplace = True)
print(df.to_string())
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
print(df.to_string())
```

```python
df.shape
```

```python
for x in df.index:
    if df.loc[x,'Duration'] > 120:
        df.loc[x,'Duration'] = 120
print(df.to_string())
```

```python
df.shape
```

```python
for x in df.index:
    if df.loc[x,'Duration'] > 60:
        df.drop(x,inplace = True )
print(df.to_string())
```
