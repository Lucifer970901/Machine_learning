How to handle the null values


delete the row containing null values
replace null values with mean of the column

```python
df.dropna(inplace = True)
print(df.to_string())
```

```python
x = df['Calories'].mean()
df.fillna({'Calories':x}, inplace = True)
```

```python
print(df.to_string())
```

```python
x = df['Calories'].mode()[0]
df.fillna({'Calories':x}, inplace = True)
print(df.to_string())
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = 'mixed')
print(df.to_string())
```

```python
df.loc[7,'Duration'] = 45
print(df.to_string())
```

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print(df.to_string())
```

```python
for x in df.index:
  if df.loc[x, "Duration"] > 60:
    df.drop(x, inplace = True)
print(df.to_string())
```

```python
df.drop_duplicates(inplace = True)
print(df.to_string())
```
