**Pandas in python** 

its a library used for data manipulation and analysis. It provides data structures like DataFrames (two-dimensional labeled arrays) and Series (one-dimensional labeled arrays) that allow for efficient handling of tabular data. 

```python
import pandas as pd
arr = [1,2,3,4,5]
myvar = pd.Series(arr)
print(myvar)
```

by default it assigns index to the values, you can change the index if you need

```python
myvar = pd.Series(arr,index = ['a','b','c','d','e'])
myvar
```

```python
print(myvar['c'])
```

**Dictionaries** 

 a built-in data structure in Python used to store collections of data in key-value pairs. Each key in a dictionary must be unique and immutable (e.g., strings, numbers, tuples), while values can be of any data type and can be duplicated. Dictionaries are mutable, which means they can be changed after they are created. 

```python
calories = {'day1':140,'day2':380,'day3':240}
print(calories)
type(calories)
```

in dictionary by default, keys become the index

```python
myvar = pd.Series(calories)
print(myvar)
```

```python
data = {
    'calories':[420,380,390],
    'duration':[50,40,45]
}
myvar = pd.DataFrame(data)
print(myvar.loc[[0,1]])
```

```python
myvar = pd.DataFrame(data, index = ['day1','day2','day3'])
print(myvar)
print(myvar.loc['day2'])
```
